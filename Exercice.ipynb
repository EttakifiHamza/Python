Date : La date exacte de chaque vente (cela peut inclure plusieurs années).

Produit : Le nom du produit vendu.

Quantité : Le nombre d'unités de produit vendues dans chaque transaction.

PrixUnitaire : Le prix de vente d'une unité du produit.

MontantVente : Le montant total de la vente, calculé en multipliant la Quantité par le PrixUnitaire.

## Q5 : Calcul de la Quantité Totale de Chaque Produit

## Q6: Afficher les quantités de chaque produit sous forme de graphique à barres

```python
import matplotlib.pyplot as plt
```

```python
# Group and Sum
grouped_data = df.groupby('Produit')['Quantite'].sum()

# Plotting as a bar chart
grouped_data.plot(kind='bar', figsize=(10, 6), title='Total Quantity by Product')
plt.ylabel('Total Quantity')
plt.xlabel('Product')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Structures de Controle

Les structures de contrôle en programmation permettent de gérer le flux d'exécution du programme, c'est-à-dire de déterminer quelles instructions seront exécutées et dans quel ordre, selon certaines conditions ou répétitions. En Python, comme dans d'autres langages de programmation, les structures de contrôle sont très utiles pour prendre des décisions, répéter des actions ou gérer des cas particuliers.

### 1. Conditionnelles (if, elif, else)

Vérification de la température pour décider de l'irrigation

Tu as une culture et tu veux décider s'il faut irriguer en fonction de la température. Le système que tu utilises prend la température en entrée et agit selon les conditions suivantes :

Si la température est supérieure à 30°C, il faut irriguer.
Si la température est supérieure à 20°C, mais inférieure ou égale à 30°C, une irrigation légère pourrait être nécessaire.
Si la température est inférieure ou égale à 20°C, il n'y a pas besoin d'irrigation.

```python
# Conditionnelle pour déterminer si l'irrigation est nécessaire en fonction de la température
temperature = 35  # Température en degrés Celsius

if temperature > 30:
    print("Il fait chaud, il faut irriguer.")
elif temperature > 20:
    print("Température moyenne, une irrigation légère pourrait être nécessaire.")
else:
    print("Température fraîche, pas besoin d'irrigation.")
```

### 2. Boucles (for, while)

Calcul de l'irrigation sur plusieurs parcelles

Dans cet exercice, nous allons calculer l'irrigation nécessaire pour plusieurs parcelles en fonction de leur besoin en eau.

Objectif :
Utiliser une liste pour représenter les différentes parcelles et une autre liste pour représenter le besoin en eau pour chaque parcelle.
Utiliser une boucle for pour afficher l'irrigation nécessaire pour chaque parcelle.

```python
# Liste des parcelles et de leur besoin en eau
parcelles = ["parcelle A", "parcelle B", "parcelle C"]
eau_par_parcelle = [100, 150, 120]  # Besoin en litres d'eau pour chaque parcelle

# Boucle 'for' pour vérifier l'irrigation nécessaire pour chaque parcelle
for i in range(len(parcelles)):
    print(f"Irrigation nécessaire pour {parcelles[i]} : {eau_par_parcelle[i]} litres.")
```

Dans cet exercice, nous utilisons une boucle while pour suivre l'absence de pluie pendant plusieurs jours. La boucle continue de tourner tant qu'il pleut, et on l'arrête après un certain nombre de jours sans pluie. Une fois ce nombre atteint, on décide qu'il faudra irriguer bientôt.

Objectif :
Utiliser une boucle while pour suivre l'absence de pluie pendant une période donnée.
Arrêter la boucle après un certain nombre de jours sans pluie, et afficher un message pour indiquer qu'il faudra irriguer.

```python
# Suivi de la pluie pendant une période
pluie_present = True
jours_sans_pluie = 0

while pluie_present:
    # On suppose qu'après chaque jour sans pluie, on vérifie à nouveau
    jours_sans_pluie += 1
    print(f"Jour {jours_sans_pluie}: Pas de pluie.")
    if jours_sans_pluie >= 3:
        pluie_present = False  # Arrête la boucle après 3 jours sans pluie

print("Trois jours sans pluie, on devra irriguer bientôt.")
```

### 3. Instruction break

Arrêter une irrigation si une certaine quantité d'eau a été atteinte

Dans cet exercice, nous contrôlons l'irrigation en vérifiant si une limite d'eau définie a été atteinte. Si cette limite est dépassée, l'irrigation s'arrête.

Objectif :


*   Simuler l'irrigation de plusieurs parcelles en ajoutant progressivement de l'eau.
*   Arrêter l'irrigation lorsque la quantité totale d'eau atteint une limite spécifiée.

```python
# Contrôler l'irrigation et arrêter si une limite est atteinte
total_eau = 0
limite_eau = 500  # Limite d'eau en litres

for i in range(5):  # 5 parcelles
    eau_necessaire = 120  # Eau nécessaire pour chaque parcelle
    total_eau += eau_necessaire
    print(f"Irrigation parcelle {i+1}, eau utilisée : {eau_necessaire} litres.")

    if total_eau >= limite_eau:
        print("Limite d'eau atteinte, arrêt de l'irrigation.")
        break
```

### 4. Instruction continue

Ignorer certaines parcelles si elles sont déjà irriguées

Dans cet exercice, on va vérifier l'état d'irrigation de chaque parcelle et ignorer celles qui ont déjà été irriguées. La boucle va seulement irriguer les parcelles qui n'ont pas encore été traitées.

Objectif :
Utiliser une boucle pour vérifier l'état de chaque parcelle (si elle a déjà été irriguée ou non).

Ignorer les parcelles irriguées et irriguer seulement celles qui ne le sont pas encore.

```python
# Liste des parcelles et leur état d'irrigation
parcelles = ["parcelle A", "parcelle B", "parcelle C"]
irrigation_effectuee = [True, False, False]

# Boucle pour irriguer uniquement les parcelles non irriguées
for i in range(len(parcelles)):
    if irrigation_effectuee[i]:
        print(f"{parcelles[i]} a déjà été irriguée. Passage à la suivante.")
        continue  # Ignore cette parcelle et passe à la suivante

    print(f"En cours d'irrigation pour {parcelles[i]}.")
```

### 5. Instruction pass

Vérifier quelque chose sans effectuer d'action spécifique

Dans cet exercice, on a une boucle qui passe en revue plusieurs jours, et on veut exécuter des actions spécifiques, sauf pour le jour 2 où aucune action n'est effectuée.

Objectif :
Utiliser l'instruction pass pour ignorer certaines conditions dans une boucle sans provoquer d'erreur.

```python
for jour in range(1, 4):
    if jour == 2:
        pass  # Pas d'action pour le jour 2
    else:
        print(f"Travail agricole effectué le jour {jour}.")
```

2.6.Fonction avec condition (logique décisionnelle)

Fonction qui prend une décision en utilisant des conditions (if, else).

Écrire une fonction stress_hydrique qui permet de détecter si une plante est en situation de stress hydrique, en fonction de la température et de l'humidité du sol.

La fonction prend deux paramètres :
temp : la température en degrés Celsius.
humidite_sol : l'humidité du sol en pourcentage.
Si la température est supérieure à 30°C et l'humidité du sol est inférieure à 40%, la fonction doit retourner "Stress hydrique élevé".
Sinon, elle retourne "Conditions normales".

```python
#détection de stress hydrique
def stress_hydrique(temp, humidite_sol):
    if temp > 30 and humidite_sol < 40:
        return "Stress hydrique élevé"
    else:
        return "Conditions normales"
print(stress_hydrique(20, 45))
```
